### About

В этой тетрадке Вам предстоит найти оптимальную конфигурацию одного слоя (view) трекинговой станции.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Geometry generation

Один слой станции (view) состоит из 4 слоев трубок. Параметры, характеризующие взаимное положение трубок, изображены на рисунке:

<img src="pic/img1.png" alt="Drawing" style="width: 300px;">

Класс **Tracker** генерирует (z, y) координаты центров трубок для задданых параметров pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane.

```python
R = 1. # cm
pitch = 3.6 #cm 
yoffset_plane = 1.6
yoffset_layer = 1.3
zshift_layer = 1.9
zshift_plane = 4.2
```

```python
class Tracker(object):
    
    def __init__(self, R, pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane):
        
        self.R = R
        self.pitch = pitch
        self.yoffset_plane = yoffset_plane
        self.yoffset_layer = yoffset_layer
        self.zshift_layer = zshift_layer
        self.zshift_plane = zshift_plane
        
        
    def create_geometry(self):
        
        base = np.arange(-100, 101, 1)
        step = self.pitch
        
        layer1_y = step * base
        layer1_z = 0. * np.ones(len(base))
        
        layer2_y = layer1_y + self.yoffset_layer
        layer2_z = layer1_z + self.zshift_layer
        
        layer3_y = layer1_y + self.yoffset_plane
        layer3_z = layer1_z + self.zshift_plane
        
        layer4_y = layer3_y + self.yoffset_layer
        layer4_z = layer3_z + self.zshift_layer
        
        Z = np.concatenate((layer1_z.reshape(-1, 1), 
                            layer2_z.reshape(-1, 1), 
                            layer3_z.reshape(-1, 1), 
                            layer4_z.reshape(-1, 1)), axis=1)
        
        Y = np.concatenate((layer1_y.reshape(-1, 1), 
                            layer2_y.reshape(-1, 1), 
                            layer3_y.reshape(-1, 1), 
                            layer4_y.reshape(-1, 1)), axis=1)
        
        geo = [Z, Y]
        
        return geo
    
```

```python
tr = Tracker(R, pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane)
geo = tr.create_geometry()
```

```python
Z, Y = geo
```

### Geometry display

Функция **geometry_display** отображает трубки по задданым (z, y) координатам трубок.

```python
def geometry_display(Z, Y, y_min=-10, y_max=10):

    Z_flat = np.ravel(Z)
    Y_flat = np.ravel(Y)

    z_min = Z_flat.min()
    z_max = Z_flat.max()

    sel = (Y_flat >= y_min) * (Y_flat < y_max)
    Z_flat = Z_flat[sel]
    Y_flat = Y_flat[sel]

    plt.figure(figsize=(6, 6 * (y_max - y_min + 2) / (z_max - z_min + 10)))
    plt.scatter(Z_flat, Y_flat)

    for z,y in zip(Z_flat, Y_flat):

        circle = plt.Circle((z, y), 1., color='b', fill=False)
        plt.gcf().gca().add_artist(circle)



    plt.xlim(z_min - 5, z_max + 5)
    plt.ylim(y_min - 1, y_max + 1)
    plt.xlabel('Z')
    plt.ylabel('Y')
    #plt.show()
```

```python
geometry_display(Z, Y, y_min=-10, y_max=10)
```

### Tracks generation

Класс **Tracks** генерирует набор прямых треков, проходящих через трубки. Трек описывается парой параметров (k, b): y = kz + b. 

```python
class Tracks(object):
    
    def __init__(self, b_min, b_max, angle_mean, angle_std):
        
        self.b_min = b_min
        self.b_max = b_max
        self.angle_mean = angle_mean
        self.angle_std = angle_std
        
    def generate(self, N):
        
        B = np.random.RandomState(42).uniform(self.b_min, self.b_max, N)
        Angles = np.random.RandomState(42).normal(self.angle_mean, self.angle_std, N)
        K = np.tan(Angles)
        
        tracks = np.concatenate((K.reshape(-1, 1), B.reshape(-1, 1)), axis=1)
        
        return tracks
```

```python
t = Tracks(-100, 100, 0, 0.2)
tracks = t.generate(1000)
```

### Tracks display

Фугкция **tracks_display** отрисовывает треки по их параметрам (k, b).

```python
def tracks_display(tracks, Z):

    Z_flat = np.ravel(Z)

    z_min = Z_flat.min()
    z_max = Z_flat.max()
    
    z1 = z_min - 5
    z2 = z_max + 5

    for k, b in tracks:
        
        plt.plot([z1, z2], [k * z1 + b, k * z2 + b], c='0.2')



    #plt.show()
```

```python
geometry_display(Z, Y, y_min=-10, y_max=10)
tracks_display(tracks, Z)
```

### Metric

Для заданной конфигурации трубок будем считать долю треков, которые пролетели через 2 и более трубок:

$$Score = \frac{N\_tracks_{n\_hits >= 2}}{N\_tracks}$$

Чем выше значение, тем лучше.

```python
def score(Z, Y, tracks, R):
    
    values = []
    
    for k, b in tracks:
        
        Y_pred = k * Z + b
        dY = np.abs(Y_pred - Y)
        
        alpha = np.arctan(k)
        cos = np.cos(alpha)
        
        is_intersect = dY * cos < R
        n_intersections = (is_intersect).sum()
        
        if n_intersections >= 2:
            values.append(1)
        else:
            values.append(0)
    
    
    return np.mean(values)
```

```python
print score(Z, Y, tracks, R)
```

## Optimization

В этой задаче необходимо найти такую конфигурацию трубок, при которой значение целевой метрики **score** будет максимально. Однако, мы будем решать задачу минимизации, поэтому необходимо минимизировать значение **1 - score**. 
Чем меньше значение новой метрики, тем лучше.

Зафиксируем радиус трубок, минимальные и максимальные значения параметров. Также, сгенерируем набор треков для вычисления значения целевой метрики.

Функция **f** по заданным парамтерам геометрии вычисляем значение целевой метрики, которую нужно минимизировать.

```python
R = 1.

t = Tracks(-100, 100, 0, 0.2)
tracks = t.generate(1000)

pitch_min, pitch_max = [3.6, 3.601]
yoffset_plane_min, yoffset_plane_max = [0, 5]
yoffset_layer_min, yoffset_layer_max = [0, 5]
zshift_layer_min, zshift_layer_max = [2, 5]
zshift_plane_min, zshift_plane_max = [4, 10]
```

```python
def f(x):
    
    pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane = x
       
    tr = Tracker(R, pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane)
    geo = tr.create_geometry()
    
    Z, Y = geo
    
    val = score(Z, Y, tracks, R)
    
    return 1. - val # Чем меньше, тем лучше.
```

### Optimization via grid search

Первый и самый простой способ оптимизации - Grid Search. В нем создается сетка параметров, и ведется перебор всем значений в сетке.

```python
%%time

n_points = 4
pitch_grid = np.linspace(pitch_min, pitch_max, n_points)
yoffset_plane_grid = np.linspace(yoffset_plane_min, yoffset_plane_max, n_points)
yoffset_layer_grid = np.linspace(yoffset_layer_min, yoffset_layer_max, n_points)
zshift_layer_grid = np.linspace(zshift_layer_min, zshift_layer_max, n_points)
zshift_plane_grid = np.linspace(zshift_plane_min, zshift_plane_max, n_points)

target_values = []
min_target_values = []
params_for_min_target_values = []

i = 0
for pitch in pitch_grid:
    for yoffset_plane in yoffset_plane_grid:
        for yoffset_layer in yoffset_layer_grid:
            for zshift_layer in zshift_layer_grid:
                for zshift_plane in zshift_plane_grid:
    
                    val = f([pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane])

                    target_values.append(val)
                    if i == 0:
                        min_target_values.append(val)
                        params_for_min_target_values.append((pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane))
                        i = 1
                    elif val < min_target_values[-1]:
                        min_target_values.append(val)
                        params_for_min_target_values.append((pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane))
                    else:
                        min_target_values.append(min_target_values[-1])
                        params_for_min_target_values.append(params_for_min_target_values[-1])
```

```python
print "Optimum score = ", min_target_values[-1]
plt.plot(min_target_values, linewidth=2)
plt.xlabel("Number of calls")
plt.ylabel('1 - Score')
plt.grid(b=1)
```

```python
(pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane) = params_for_min_target_values[-1]
print "Optimal parameters: "
print "pitch = ", pitch
print "yoffset_plane = ", yoffset_plane
print "yoffset_layer = ", yoffset_layer
print "zshift_layer = ", zshift_layer
print "zshift_plane = ", zshift_plane
```

```python
tr = Tracker(R, pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane)
geo = tr.create_geometry()
Z, Y = geo
geometry_display(Z, Y, y_min=-10, y_max=10)
```

### Optimization via random points

В этом методе на всем диапазоне значений параметров генерируются N точек, в которых вычисляется значение целевой метрики. Точка с наилучшим значением метрики считается оптимумом.

```python
%%time

n_points = 1000
seed = None

pitch_grid = np.random.RandomState(11).uniform(pitch_min, pitch_max, n_points)
yoffset_plane_grid = np.random.RandomState(12).uniform(yoffset_plane_min, yoffset_plane_max, n_points)
yoffset_layer_grid = np.random.RandomState(13).uniform(yoffset_layer_min, yoffset_layer_max, n_points)
zshift_layer_grid = np.random.RandomState(14).uniform(zshift_layer_min, zshift_layer_max, n_points)
zshift_plane_grid = np.random.RandomState(15).uniform(zshift_plane_min, zshift_plane_max, n_points)

target_values = []
min_target_values = []
params_for_min_target_values = []

for i in range(n_points):
    
    pitch = pitch_grid[i]
    yoffset_plane = yoffset_plane_grid[i]
    yoffset_layer = yoffset_layer_grid[i]
    zshift_layer = zshift_layer_grid[i]
    zshift_plane = zshift_plane_grid[i]
    
    val = f([pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane])

    target_values.append(val)
    if i == 0:
        min_target_values.append(val)
        params_for_min_target_values.append((pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane))
    elif val < min_target_values[-1]:
        min_target_values.append(val)
        params_for_min_target_values.append((pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane))
    else:
        min_target_values.append(min_target_values[-1])
        params_for_min_target_values.append(params_for_min_target_values[-1])
```

```python
print "Optimum score = ", min_target_values[-1]
plt.plot(min_target_values, linewidth=2)
plt.xlabel("Number of calls")
plt.ylabel('1 - Score')
plt.grid(b=1)
```

```python
(pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane) = params_for_min_target_values[-1]
print "Optimal parameters: "
print "pitch = ", pitch
print "yoffset_plane = ", yoffset_plane
print "yoffset_layer = ", yoffset_layer
print "zshift_layer = ", zshift_layer
print "zshift_plane = ", zshift_plane
```

```python
tr = Tracker(R, pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane)
geo = tr.create_geometry()
Z, Y = geo
geometry_display(Z, Y, y_min=-10, y_max=10)
```

### Optimization via Bayesian methods

Третий метод оптимизации используем Гауссовские процессы.

```python
from skopt import gp_minimize

kappa= 10.1

res = gp_minimize(func=f,                  # the function to minimize
                  dimensions=[(pitch_min, pitch_max), 
                              (yoffset_plane_min, yoffset_plane_max), 
                              (yoffset_layer_min, yoffset_layer_max), 
                              (zshift_layer_min, zshift_layer_max), 
                              (zshift_plane_min, zshift_plane_max)],      # the bounds on each dimension of x
                  acq_func="LCB",      # the acquisition function
                  n_calls=100,         # the number of evaluations of f 
                  n_random_starts=50,  # the number of random initialization points
                  noise=0.01**2,       # the noise level (optional)
                  random_state=123, 
                  kappa=kappa, 
                  n_jobs=3)   # the random seed
```

```python
from skopt.plots import plot_convergence
plot_convergence(res);
```

```python
pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane = res.x
print "Optimal parameters: "
print "pitch = ", pitch
print "yoffset_plane = ", yoffset_plane
print "yoffset_layer = ", yoffset_layer
print "zshift_layer = ", zshift_layer
print "zshift_plane = ", zshift_plane
```

```python
tr = Tracker(R, pitch, yoffset_plane, yoffset_layer, zshift_layer, zshift_plane)
geo = tr.create_geometry()
Z, Y = geo
geometry_display(Z, Y, y_min=-10, y_max=10)
```
